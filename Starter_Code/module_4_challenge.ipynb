## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
df.columns
```

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
#.apply(lambda x: '{:.2f}'.format(x))
df['Subtotal'] = (df['unit_price'] * df['qty'])
df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
def shipping_price(column):
    if (column['unit_weight'] * column['qty']) > 50:
        return (column['unit_weight'] * column['qty']) * 7
    return (column['unit_weight'] * column['qty']) * 10
    
df['shipping_price'] = df.apply(shipping_price, axis=1)
df.head()
```

```python
df['qty'].max()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

def calculate_total(row):
    if 'Subtotal' in row and 'shipping_price' in row:
        subtotal = float(row['Subtotal'])
        shipping_price = float(row['shipping_price'])
        total = (subtotal + shipping_price) * 1.0925
    return '{:.2f}'.format(total)   
    #x=0
    #if (columns[x]):
        #total = (columns[str('Subtotal')] + columns['shipping_price']) * 0.0925
        #return '{:.2f}'.format(total)

df['total_price'] = df.apply(calculate_total, axis=1)
df.head()
```

```python
#Saving code for formatting decimal points for reference
# '{:.2f}'.format
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
def calculate_cost(row):
    if 'unit_cost' in row:
        unit_cost = float(row['unit_cost'])
        qty = float(row['qty'])
        shipping_price = float(row['shipping_price'])
        true_cost = (unit_cost * qty) + shipping_price
    return true_cost
    
df['true_cost'] = df.apply(calculate_cost, axis=1)
df.head()
```

```python
# Create a column for the profit of
# each line using line cost and line price
def calculate_profit (row):
    try:
        if 'unit_price' in row:
            line_cost = float(row['unit_cost'])
            line_price = float(row['unit_price'])
            qty = float(row['qty'])
            line_profit = (line_price - line_cost) * qty
            return line_profit
    except (ValueError, TypeError):
        return '0.00'

df['line_profit'] = df.apply(calculate_profit, axis=1)
df.head()
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
summary_df = summary_df.rename(columns={
                               'client_id' : 'Client ID',
                               'qty' : 'Units',
                               'shipping_price' : 'Shipping',
                               'total_price': 'Total Revenue',
                               'true_cost' : 'Total Cost',
                               'line_profit' : 'Total Profit'})


summary_df.head()
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
summary_df_profit = summary_df.sort_values(by='Total Profit', ascending=False)
summary_df_profit.head()
```

```python
#Formatting the numbers to be in the millions
def make_millions(x):
    return '{:.2f}M'.format(x/1000000)

summary_df_profit['Shipping'] = summary_df_profit['Shipping'].apply(make_millions)
summary_df_profit['Total Revenue'] = summary_df_profit['Total Revenue'].apply(make_millions)
summary_df_profit['Total Cost'] = summary_df_profit['Total Cost'].apply(make_millions)
summary_df_profit['Total Profit'] = summary_df_profit['Total Profit'].apply(make_millions)

summary_df_profit.head()
```
